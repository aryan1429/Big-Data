```python
#Create
hbase_table = {}
```

```python
row_key = "row1"
```

```python
hbase_table[row_key] = {
    "name" : "Aryan",
    "age" : 69,
    "city" : "hyd"
}
```

```python
print("After Insert",hbase_table)
```

```python
#Read
row = hbase_table.get(row_key)
print("Read Row",row)
```

```python
#Update
hbase_table[row_key]["age"] = 21
print("Ater Update",hbase_table)
```

```python
#Delete
del hbase_table[row_key]
print("After Delete:",hbase_table)
```
