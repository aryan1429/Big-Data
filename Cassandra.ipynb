```python
import pandas as pd
```

```python
data = {
    "id" : [1,2,3,4],
    "name" : ["Aryan","Alejandro","Aligeti","Supremo"],
    "age" : [21,22,23,24],
    "city" : ["HYD","KOL","DEL","MUM"]
}
```

```python
df = pd.DataFrame(data)
print("Initial Table \n",df)
```

```python
#Create(Insert)
new_row = pd.DataFrame({
    "id" : [5],
    "name" : ["Aryan"],
    "age" : [21],
    "city" : ["Phil"]
})

df = pd.concat([df,new_row], ignore_index=True)
print("After Insert\n",df)
```

```python
#Read
row = df[df["id"] == 2]
print(row)
```

```python
#Update
df.loc[df["id"] == 2, "age"] = 69
print(df)
```

```python
#delete
df = df[df["id"] != 1]
print(df)
```

```python
#data analysis
avg_age = df["age"].mean()
print("Average Age:",avg_age)
```

```python
count_by_city = df["city"].value_counts()
print(count_by_city)
```
